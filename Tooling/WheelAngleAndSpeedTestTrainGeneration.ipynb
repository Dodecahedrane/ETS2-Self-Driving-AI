```python
import cv2
import numpy as np
from PIL import Image, ImageDraw, ImageFilter, ImageEnhance
import os
import shutil
```

```python
def Setup(WheelPath, SpeedPath):
    path = "WheelAngleLabel"
    exist = os.path.exists(path)
    if not exist:
           os.makedirs(path)
    else:
        print('Directory Exists')
        # Uncomment these two lines if you want to delete the frames folder before adding more frames
        shutil.rmtree(path)
        os.makedirs(path)
        
    path = "WheelAngleLabel/Angle"
    exist = os.path.exists(path)
    if not exist:
           os.makedirs(path)
    else:
        print('Directory Exists')
        # Uncomment these two lines if you want to delete the frames folder before adding more frames
        shutil.rmtree(path)
        os.makedirs(path)
    
    path = "WheelAngleLabel/NoAngle"
    exist = os.path.exists(path)
    if not exist:
           os.makedirs(path)
    else:
        print('Directory Exists')
        # Uncomment these two lines if you want to delete the frames folder before adding more frames
        shutil.rmtree(path)
        os.makedirs(path)
    
    path = "WheelAngleLabel/Original"
    exist = os.path.exists(path)
    if not exist:
           os.makedirs(path)
    else:
        print('Directory Exists')
        # Uncomment these two lines if you want to delete the frames folder before adding more frames
        shutil.rmtree(path)
        os.makedirs(path)
    
    
    
    
    path = "SpeedLabel"
    exist = os.path.exists(path)
    if not exist:
           os.makedirs(path)
    else:
        print('Directory Exists')
        # Uncomment these two lines if you want to delete the frames folder before adding more frames
        shutil.rmtree(path)
        os.makedirs(path)
```

```python
def WheelImages(image, path):
    #load image
    img = cv2.imread(image)
    
    #dont want the '.jpg' element of the file name in the file name
    justFile = image.split('/')[1]
    noExtension = justFile[:-4]
    
    #crop to steering wheel
    crop = img[600:1080, 590:1350]
    #save crop
    cv2.imwrite('img.jpg',crop)
    
    #load background and foreground
    background = Image.open("img.jpg")
    foreground = Image.open("anglePink.png")
    
    #save the image unmodified
    background.save(path + '/Original/' + noExtension + '.jpg', quality=95)
    
    #brighten the image
    enhancer = ImageEnhance.Brightness(background)
    factor = 4
    background = enhancer.enhance(factor)
    
    #scale the angle image
    scale = 1.5
    h = int(abs(600*scale))
    w = int(abs(300*scale))
    foreground = foreground.resize([h,w],Image.Resampling.LANCZOS)    
    
    #save the image without the angle
    background.save(path + '/NoAngle/' + noExtension + '.jpg', quality=95)
    
    #composite and save image
    background.paste(foreground, (-65, -25), foreground)
    background.save(path + '/Angle/' + noExtension + '.jpg', quality=95)
```

```python
def SpeedoImage(image, path):
    #load image
    img = cv2.imread(image)
    
    #dont want the '.jpg' element of the file name in the file name
    justFile = image.split('/')[1]
    noExtension = justFile[:-4]

    #crop to steering wheel
    crop = img[711:728, 54:75]
    #save crop
    cv2.imwrite(path + '/' + noExtension + 'Speed' + '.jpg',crop)
```

```python
def GenerateImageForLabeling(image, wheelPath, speedPath):
    WheelImages(image, wheelPath)
    SpeedoImage(image, speedPath)
```

```python
Setup("WheelAngleLabel", "SpeedLabel")

directory = os.fsencode('Frames')

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".jpg"):
        #print(os.path.join(filename))
        GenerateImageForLabeling('Frames/' + filename, "WheelAngleLabel", "SpeedLabel")
```
