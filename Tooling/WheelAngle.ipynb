```python
import cv2

import numpy as np
from matplotlib import pyplot as plt
```

```python
path = '/Examples/'
images = ['Example/day1.jpg','Example/day2.jpg','Example/day3.jpg','Example/day4.jpg',
          'Example/night1.jpg','Example/night2.jpg']
```

```python
def ShowImg(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(gray)
    plt.title('my picture')
    plt.show()
```

```python
def increase_brightness(img, value=30):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    return img
```

```python
def WheelImageCrop(frame):    
    #crop to steering wheel
    frameCrop = frame[600:1080, 590:1350]
    
    brighter = increase_brightness(frameCrop, value=50)
    
    
    
    
    return brighter
```

```python
img = cv2.imread(images[0])
imgCrop = WheelImageCrop(img)
ShowImg(imgCrop)
```
